```python
df['Education_Level'] = df['Education_Level'].replace('Unknown' , np.nan)
print(df['Education_Level'].isnull().sum())
```

```python
value_counts = df['Education_Level'].value_counts()

print(value_counts)
plt.figure(figsize=(8, 5))
value_counts.plot(kind='pie', colors=['skyblue','Yellow','Red','Blue','brown','black'])
```
